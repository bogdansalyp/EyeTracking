# Data Preprocessing

```python
from mtcnn import MTCNN
from PIL import Image
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
```

```python
DIRECTORY = 'dataDir/'
PREPROCESSED_DIR = 'preprocessedData/'
size = 128, 128
```

```python
detector = MTCNN()
```

```python
filenum = 0
for dirname in os.listdir(DIRECTORY):
    if dirname != '.DS_Store':
        for filename in tqdm(os.listdir(os.path.join(DIRECTORY, dirname))):
            if filename.endswith(".jpg"):
                pixels = plt.imread(os.path.join(DIRECTORY, dirname, filename))
                
                # detect face
                results = detector.detect_faces(pixels)
                x1, y1, width, height = results[0]['box']
                x2, y2 = x1 + width, y1 + height
                face = pixels[y1:y2, x1:x2]
                
                # resize
                im = Image.fromarray(face)
                im.thumbnail(size, Image.ANTIALIAS)
                plt.imshow(im)
                
                # detect side of a screen
                result_class = int(int(filename.split('{')[1].split(':')[0]) > 550)
#                 print(result_class)
                filenum += 1
                
                im.save(os.path.join(PREPROCESSED_DIR, str(result_class), '{}.jpg'.format(filenum)))
                #print(os.path.join(DIRECTORY, dirname, filename))
```
